```python
from PIL import Image
import numpy as np
import random
from matplotlib import pyplot as plt

path = "XDD.jpg"
mtx = Image.open(path).convert('1')
mtx = np.array(mtx)

secret1 = np.array([[0 for _ in range(len(mtx[0])*2)] for _ in range(len(mtx))])
secret2 = np.array([[0 for _ in range(len(mtx[0])*2)] for _ in range(len(mtx))])

for i in range(len(mtx)):
    for j in range(len(mtx[0])):
        rand = random.randint(0, 1)
        secret1[i][j*2] = not rand
        secret1[i][j*2+1] = rand
        secret2[i][j*2] = (rand + mtx[i][j])%2
        secret2[i][j*2+1] = not secret2[i][j*2]

display(Image.fromarray(mtx))
plt.imshow(secret1, cmap="gray")
plt.show()
plt.imshow(secret2, cmap="gray")
plt.show()
plt.imshow(secret1+secret2, cmap="gray")
plt.show()
```
